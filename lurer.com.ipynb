```python
import re 
import time
import requests
import numpy as np
import pandas as pd
from textblob import TextBlob , Word
from scrapy.http import TextResponse
import nltk
from nltk.corpus import stopwords , wordnet
from nltk.tokenize import sent_tokenize , word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
url = "https://lurer.com/?l=en"
page = requests.get(url)
response = TextResponse(url= page.url, body = page.text , encoding = "utf-8")
```

```python
titles = response.css("a[class='clearfix']::text").extract()
                                    
```

```python
titles
```

```python
dates = response.css("a[class='clearfix']>time::text").re("\d\d:\d\d")
```

```python
dates
```

```python
titles_1 = TextBlob(" ".join(titles))
```

```python
titles_1.words
```

```python
sw = stopwords.words('english')
```

```python
sw_titles = [i for i in titles_1.words.lower() if i not in sw]
```

```python
df = pd.DataFrame(sw_titles)
```

```python
df
```

```python
popular = df[0].value_counts()
```
