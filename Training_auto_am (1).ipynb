```python
def cars_scraper_year(url):
    page = browser.page_source
    response = TextResponse(url=url,body = page,encoding = "utf-8")
    cars = response.css("span[class='grey-text']::text").extract()
    return cars
```

```python
all_cars_years = []
i = 0
while i in range(len(all_urls)) :
    all_cars_years.extend(cars_scraper_year(all_urls[i]))
    i = i +1
    time.sleep(2)
```

```python
len(all_cars_years)
```

```python
def price_scraper(url):
    page = browser.page_source
    response = TextResponse(url=url,body = page,encoding = "utf-8")
    prices_dollar = response.css("div[class='ad-mob-price bold']>span::text").extract()
    return prices_dollar
```

```python
all_prices = []
i = 0
while i in range(len(all_urls)) :
    all_prices.extend(price_scraper(all_urls[i]))
    i = i +1
    time.sleep(2)
```
